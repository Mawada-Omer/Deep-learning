```python
df = pd.read_csv('/home/ahmed/Downloads/pima-indians-diabetes.csv', header  =None)
```

```python
df.columns = ['No Preg', 'glucose', 'bp mmHg', 'skin thick mm', 'serum insulin muU/ml', 'bmi kg/m2', 'Diabetes Pedigree func', 'Age', 'Target']
```

```python
X = df.iloc[:,0:8]
X.head(3)
```

```python
y = df.iloc[:, -1]
y.head(4)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(16, input_dim=8, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
#compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the keras model on the dataset
model.fit(X, y, epochs=250, batch_size=10)
```

```python
# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
ypred = model.predict(X)
ypred.round(2)
```
